```python
import nltk
import re

nltk.download('punkt_tab')

text = """Beans. I was trying to explain to somebody as we were flying in, that's corn.
         That's beans. And they were very impressed at my agricultural knowledge.
         Please give it up for Amaury once again for that outstanding introduction.
         I have a bunch of good friends here today, including somebody who I served with,
         who is one of the finest senators in the country, and we're lucky to have him,
         your Senator, Dick Durbin is here. I also noticed, by the way,
         former Governor Edgar here, who I haven't seen in a long time, and
         somehow he has not aged and I have. And it's great to see you, Governor.
         I want to thank President Killeen and everybody at the U of I System for
         making it possible for me to be here today. And I am deeply honored at the Paul
         Douglas Award that is being given to me. He is somebody who set the path for so
         much outstanding public service here in Illinois. Now, I want to start by
         addressing the elephant in the room. I know people are still wondering why
         I didn't speak at the commencement."""

dataset = nltk.sent_tokenize(text)

for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] = re.sub(r'\W', ' ', dataset[i])
    dataset[i] = re.sub(r'\s+', ' ', dataset[i])

for i, sentence in enumerate(dataset):
    print(f"Sentence {i+1}: {sentence}")
```

```python
import nltk
from nltk.corpus import stopwords
import pandas as pd
nltk.download('stopwords')

word2count = {}

for data in dataset:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count:
            word2count[word] = 1
        else:
            word2count[word] += 1

stop_words = set(stopwords.words('english'))

filtered_word2count = {word: count for word, count in word2count.items() if word not in stop_words}

word_freq_df = pd.DataFrame(list(filtered_word2count.items()), columns=['Word', 'Frequency'])

word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

print(word_freq_df)
```

```python
import heapq
import matplotlib.pyplot as plt

freq_words = heapq.nlargest(10, word2count, key=word2count.get)

print(f"Top 10 frequent words: {freq_words}")

top_words = sorted(word2count.items(), key=lambda x: x[1], reverse=True)[:10]
words, counts = zip(*top_words)

plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xticks(rotation=45)
plt.title('Top 10 Most Frequent Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import seaborn as sns

X = []

for data in dataset:
    vector = []
    for word in freq_words:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        else:
            vector.append(0)
    X.append(vector)

X = np.asarray(X)

plt.figure(figsize=(10, 6))
sns.heatmap(X, cmap='RdYlGn', cbar=False, annot=True, fmt="d", xticklabels=freq_words, yticklabels=[f"Sentence {i+1}" for i in range(len(dataset))])

plt.title('Bag of Words Matrix')
plt.xlabel('Frequent Words')
plt.ylabel('Sentences')
plt.show()
```

## Text Generation with Hugging Face Transformers

Text generation is a Natural Language Processing (NLP) task where a model produces new text based on a given prompt or context. Transformer models, especially those based on architectures like GPT (Generative Pre-trained Transformer), excel at this task due to their ability to understand and generate human-like text.

We'll use the `pipeline` function from the `transformers` library, which simplifies using pre-trained models for various NLP tasks, including text generation.

```python
from transformers import pipeline

# Initialize the text generation pipeline
# We'll use 'distilgpt2' which is a smaller, faster version of GPT-2
# suitable for quick demonstrations.
text_generator = pipeline("text-generation", model="distilgpt2")

# Define a prompt (the starting text for generation)
prompt = "Once upon a time, in a land far, far away,"

# Generate text
# You can adjust parameters like max_length, num_return_sequences, and temperature
# max_length: The maximum number of tokens to generate.
# num_return_sequences: The number of different sequences to generate.
# temperature: Controls the randomness of the generation (higher = more random).

generated_text = text_generator(
    prompt,
    max_length=50,
    num_return_sequences=1,
    truncation=True # Important for handling longer prompts
)

# Print the generated text
print("Generated Text:")
for i, output in enumerate(generated_text):
    print(f"--- Generated Sequence {i+1} ---")
    print(output['generated_text'])
    print("\n")
```

```python
## Code Generation with NLP Transformers: A Project Overview

Code generation is a specialized field of NLP where models are trained to produce source code in various programming languages based on natural language descriptions or existing code contexts. While general-purpose language models like `distilgpt2` (which we've been using for text generation) can attempt to generate code, they are not optimized for it and often produce syntactically incorrect or logically flawed results.

For effective code generation, models explicitly trained on vast datasets of source code are used. Some prominent examples include:

*   **CodeGPT**
*   **CodeBERTa**
*   **CodeLlama**
*   **AlphaCode**
*   **OpenAI Codex (and models like GPT-3/GPT-4 for programming tasks)**

These models understand programming language syntax, common algorithms, and often have a better grasp of logical flow. Setting up and using these larger, specialized models typically involves installing specific libraries and potentially accessing models through APIs or downloading large weights, which might require more resources.

### Demonstrating Code Generation (with `distilgpt2` limitations)

For the purpose of demonstration within our current environment, we will use `distilgpt2` to attempt generating a Python function. Please be aware that the quality will be limited as `distilgpt2` was not fine-tuned for code.
```

```python
from transformers import pipeline

# Re-initialize the text generation pipeline if needed, using distilgpt2
# (If you want to use a specialized code model, you would load it here instead)
text_generator = pipeline("text-generation", model="distilgpt2")

# Define a detailed prompt to guide code generation for a Python function
code_prompt = '''# Python function to calculate the factorial of a number.
# The function should take an integer n as input.
# It should return 1 if n is 0, and n multiplied by factorial of (n-1) otherwise.
def factorial(n):
    """Calculates the factorial of a non-negative integer."""
'''

print("Attempting to generate Python code using distilgpt2...")

# Generate code
# We'll set a higher max_length to allow for the function body.
# temperature is kept balanced to encourage some structure.
generated_code_output = text_generator(
    code_prompt,
    max_length=150, # Increased length for function body
    num_return_sequences=1,
    temperature=0.6, # Adjusted for a balance between creativity and structure
    truncation=True
)

# Extract and print the generated code
generated_code_str = generated_code_output[0]['generated_text']
print("\n--- Generated Python Code (distilgpt2) ---")
print(generated_code_str)

# --- Optional: Attempt to execute the generated code (USE WITH CAUTION) ---
# This part demonstrates how you might integrate generated code into a workflow.
# It's crucial to review and sanitize any generated code before execution in a real project.
print("\n--- Attempting to execute generated code ---")
try:
    # Using a dictionary for local scope to execute the code
    exec_locals = {}
    exec(generated_code_str, globals(), exec_locals)

    # Try to call the generated function if it exists
    if 'factorial' in exec_locals:
        print("Generated 'factorial' function found. Testing...")
        test_number = 5
        result = exec_locals['factorial'](test_number)
        print(f"Factorial of {test_number} (generated function): {result}")
    else:
        print("No 'factorial' function found in generated code.")

except Exception as e:
    print(f"Error executing generated code: {e}")
    print("This is expected for general models not fine-tuned for code correctness.")

print("\n--- End of Code Generation Demonstration ---")
print("For production-level code generation, explore models like CodeGPT, CodeLlama, or AlphaCode.")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load pretrained transformer chatbot model
model_name = "microsoft/DialoGPT-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print("Chatbot is ready! Type 'quit' to stop.")

chat_history_ids = None

while True:

    user_input = input("You: ")

    if user_input.lower() == "quit":
        break

    # Encode user input
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Append conversation history
    bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

    # Generate response using transformer
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode response
    bot_response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    print("Bot:", bot_response)
```
